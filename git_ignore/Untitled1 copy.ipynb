```python
import pandas as pd
import numpy as np
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
#import get to call a get request on the site
from requests import get
```

```python
#get the first page of motorcyle listings
response = get('https://newyork.craigslist.org/d/motorcycles-scooters/search/mca')
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(response.content)
```

```python
results = html_soup.find(class_='rows')
print(results.prettify())
```

```python
bike_elements = results.find_all('li', class_ ='result-row')
for bike_elem in bike_elements:
#         print(bike_elem.text)
    price_elem =bike_elem.find('span', class_='result-price')
    url_elem =bike_elem.find('a', class_='result-image gallery')['href']
    title_elem =bike_elem.find('a', class_="result-title hdrlnk")
    print(title_elem.text.strip())
    print(url_elem)
    print(price_elem.text.strip())
    print()
```

```python
response = get('https://newyork.craigslist.org/que/mcy/d/college-point-ktm-1190-adventure/7392509257.html')
html_soup = BeautifulSoup(response.text, 'html.parser')

```

```python
attributes = html_soup.find_all('p', class_='attrgroup')
```

```python
for attribute in attributes:
    spans = attribute.find_all('span')
    for span in spans:
        text = span.text.strip()
        print(text)
```
